6(a) Please randomly choose two sets of nodes in Facebook-Ego dataset (each consists of 50 nodes
without repetition), and calculate the cut size and conductance between the two sets of nodes.
Please also plot the partitioning

````python
# 50 nodes for the network (Facebook dataset)
random.seed(123)
set1_FE = random.sample(FE_G.nodes(),50) # select 50 random nodes from the facebook-ego dataset
left_nodes = [i for i in FE_G.nodes() if i not in set1_FE]
set2_FE = random.sample(left_nodes,50)
print(set(set1_FE))
print(set(set2_FE))
print(nx.cut_size(FE_G, set(set1_FE), set(set2_FE)))
nx.conductance(FE_G, set(set1_FE), set(set2_FE))
````

````python
connection = []
for u in set1_FE:
  for v in set2_FE:
    if FE_G.has_edge(u, v):
      connection.append([u,v])
print(connection)
````

````python
from networkx.drawing.nx_pydot import graphviz_layout
gw1 = FE_G.subgraph(set1_FE)
#pos1 = nx.nx_agraph.graphviz_layout(gw1,prog='dot')

pos1 = graphviz_layout(gw1, prog = "twopi")

gw2 = FE_G.subgraph(set2_FE)
pos2 = graphviz_layout(gw2,prog="twopi")

shift = 1000
for key in pos2:
    pos2[key] = (pos2[key][0]+shift, pos2[key][1])
# Combine the graphs and remove all edges

g12 = nx.disjoint_union(gw1,gw2)
for i,j in g12.edges():
    g12.remove_edge(i,j)

# Add the conjoined edges
for edge in connection:
  g12.add_edge(edge[0], edge[1])

pos12 = pos1.copy()
for node in pos2:
    pos12[node] = pos2[node]

nx.draw(gw1, pos1, node_size=40, node_color='red', vmin=0.0, vmax=1.0, with_labels=False)
nx.draw(gw2, pos2, node_size=40, node_color='blue', vmin=0.0, vmax=1.0, with_labels=False)
nx.draw_networkx_edges(g12,pos=pos12,width=1, edge_color='green', style = 'dashdot' )

plt.axis('off')
plt.show()


````
